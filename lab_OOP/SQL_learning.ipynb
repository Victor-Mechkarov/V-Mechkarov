```python
import pandas as pd
import sqlite3


# 1. Създаваме връзка с база данни (ако файлът не съществува, ще бъде създаден)
conn = sqlite3.connect('moiata_baza.db')
cursor = conn.cursor()

# 2. Създаваме таблица "potrebiteli"
cursor.execute('''
    CREATE TABLE IF NOT EXISTS potrebiteli (
        id INTEGER PRIMARY KEY,
        ime TEXT NOT NULL,
        grad TEXT
    )
''')

# 3. Добавяме малко тестови данни
cursor.execute("INSERT INTO potrebiteli (ime, grad) VALUES ('Иван', 'София')")
cursor.execute("INSERT INTO potrebiteli (ime, grad) VALUES ('Мария', 'Пловдив')")
conn.commit()

# 4. Използваме SQL заявка, за да прочетем данните чрез Pandas (за по-красив изглед)
df = pd.read_sql_query("SELECT * FROM potrebiteli", conn)
print(df)

# Затваряме връзката
conn.close()
```
