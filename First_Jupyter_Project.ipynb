```python
from collections import Counter

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.random.randint(1, 8, 20)
```

```python
results
```

```python
plt.plot(results, color='b')
plt.grid(False)
plt.xlabel("Number of experiments")
plt.ylabel("Outcomes")
plt.show()
```

```python
counts = Counter(results)
```

```python
counts.keys()
```

```python
plt.plot(counts.keys(), counts.values(), color='b')
plt.xlabel("Number of experiments")
plt.ylabel("Count")
plt.grid(False)
plt.show()
```

```python
plt.scatter(counts.keys(), counts.values())
plt.grid(False)
plt.show()
```

```python
for key, value in counts.items():
    print(f"{key} : {value}")
```

```python
plt.plot(counts.values(),counts.keys() ,color='g', linestyle='solid')
plt.grid(False)
plt.show()
```

```python
value_list = []
for value in counts.values():
    value_list.append(value)
```

```python
value_list
```

```python
for i in value_list:
    i+=1
    print(i)
```
