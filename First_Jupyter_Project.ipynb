```python
from collections import Counter

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.random.randint(1, 8, 30)
```

```python
results
```

```python
plt.plot(results, color='b')
plt.grid(False)
plt.xlabel("Number of experiments")
plt.ylabel("Outcomes")
plt.show()
```

```python
counts = Counter(results)
```

```python
counts.keys()
```

```python
plt.bar(counts.keys(), counts.values(), color='b')
plt.xlabel("Number of experiments")
plt.ylabel("Count")
plt.grid(False)
plt.show()
```
