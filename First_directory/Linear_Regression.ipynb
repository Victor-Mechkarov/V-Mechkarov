```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python


# Data (equal to 15 elements)
X = np.array([595, 520, 715, 405, 680, 490, 565, 580, 615, 435, 440, 515, 380, 510, 565]).reshape(-1, 1)
y = np.array([68, 55, 65, 42, 42, 64, 45, 56, 59, 56, 42, 38, 50, 37, 42])
```

```python

# Creating the model
model = LinearRegression()
model.fit(X, y)

# Forecast of regression line
y_pred = model.predict(X)
```

```python

# Visualisation via Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Real data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression line (Forecast)')
plt.title('Linear regression: IQ Test vs Test Data')
plt.xlabel('IQ Test Score')
plt.ylabel('Test Data Result')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Example forecast for ( IQ = 600)
new_iq = np.array([[600]])
prediction = model.predict(new_iq)
print(f"Prediction for IQ 600: {prediction[0]:.2f}")
```
